```python
import h5py
import numpy as np
from pzest.config import load_config
from pzest.preprocessing import preprocess
```

Load config and run preprocess

```python
config = load_config("../config/default.yaml")
```

```python
preprocess(config)
```

Verify output

```python
with h5py.File(config.paths.processed_hdf5_file, "r") as f:
    print("Keys:", list(f.keys()))
    for key in f.keys():
        obj = f[key]
        print(f"{key}: shape={obj.shape}, dtype={obj.dtype}")

    images = f["images"][:10]
    redshifts = f["redshift"][:]

    print(f"\nImage value range: [{images.min():.3f}, {images.max():.3f}]")
    print(f"Image mean: {images.mean():.3f}")
    print(f"Redshift range: [{redshifts.min():.3f}, {redshifts.max():.3f}]")
    print(f"Any NaN in images: {np.isnan(images).any()}")
    print(f"Any NaN in redshifts: {np.isnan(redshifts).any()}")
```
